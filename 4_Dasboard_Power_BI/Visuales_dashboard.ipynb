Esta imagen muestra el modelo de datos en Power BI, donde se evidencia la relación entre las tablas originales (clientes, ventas, productos, detalle de ventas y calendario) y la tabla central general. # **Vista general del dashboard**

![Vista](../4_Dasboard_Power_BI/imagenes/Vista.png)

Esta pantalla muestra el dashboard interactivo final, donde se visualizan los principales KPIs del comercio Aurelion, incluyendo ventas totales, número de ventas y ticket promedio. Además, se presentan análisis por categoría de producto (alimentos y limpieza), evolución temporal de las ventas, medios de pago, clientes, ciudades y productos. Los valores reflejan los datos ya corregidos tras la recategorización de productos.

## **Medidas creadas con DAX**

![Medidas](../4_Dasboard_Power_BI/imagenes/Medidas.png)

En esta captura se observa el conjunto de medidas DAX creadas para el cálculo de métricas clave como ventas totales, ventas mensuales, crecimiento mensual, clientes activos, ticket promedio y ventas máximas y mínimas. El uso de medidas permite que los KPIs y visualizaciones respondan dinámicamente a los filtros del dashboard.


## **Tabla general consolidada (Power Query)**

![Jerarquia](../4_Dasboard_Power_BI/imagenes/General.png)

En esta captura se visualiza la tabla general, creada mediante la combinación de consultas en Power Query. La tabla consolida las columnas necesarias de clientes, productos y ventas, eliminando redundancias y funcionando como una tabla de hechos que sirve de base para el modelado, las medidas DAX y las visualizaciones del dashboard.
