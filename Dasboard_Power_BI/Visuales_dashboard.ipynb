## **Modelo de datos**

![Modelo](../Dasboard_Power_BI/imagenes/Modelo.png) 

Esta imagen muestra el **modelo de datos del dashboard**, construido a partir de cuatro archivos CSV: clientes, ventas, detalle de ventas y productos. El modelo sigue un enfoque relacional, donde la tabla de ventas actúa como eje central y se relaciona con las demás tablas mediante claves únicas.
Adicionalmente, se incluye una **tabla de calendario**, utilizada para soportar el análisis temporal. Este diseño permite que las métricas respondan correctamente a los filtros y segmentadores del dashboard, asegurando consistencia y coherencia en los resultados.

# **Vista general del dashboard**

![Vista](../Dasboard_Power_BI/imagenes/Vista.png)

Esta captura presenta la **vista general del dashboard**, donde se integran los principales KPIs y visualizaciones. En la parte superior se muestran indicadores clave como ventas totales, número de ventas y ticket promedio, proporcionando una visión ejecutiva inmediata del desempeño comercial.
El dashboard combina análisis temporal, segmentación por ciudad, medio de pago, clientes y categorías de producto, permitiendo obtener una visión global del negocio desde distintos enfoques.

## **Medidas creadas con DAX**

![Medidas](../Dasboard_Power_BI/imagenes/Medidas.png)

En esta imagen se observan las **medidas desarrolladas utilizando DAX**, empleadas para el cálculo de los indicadores principales del dashboard. Estas medidas incluyen métricas de volumen, valor y análisis temporal, como ventas totales, ticket promedio, ventas mensuales, crecimiento porcentual y valores extremos.
El uso de medidas, en lugar de columnas calculadas, permite que los resultados se ajusten dinámicamente al contexto de filtros y segmentadores, habilitando un análisis interactivo y flexible.


## **Jerarquías temporales**

![Jerarquia](../Dasboard_Power_BI/imagenes/Jerarquia.png)

Esta captura muestra la configuración de **jerarquías temporales** basadas en la tabla de calendario. Las jerarquías incluyen niveles como año, trimestre y mes, lo que facilita la navegación y el análisis de la información a distintas escalas de tiempo.
Gracias a estas jerarquías, el usuario puede explorar el comportamiento de las ventas de forma progresiva, desde una visión anual hasta un análisis mensual más detallado.
